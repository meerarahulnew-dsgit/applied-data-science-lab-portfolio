```python
# 1. Import libraries
import pandas as pd
import numpy as np

# 2. Load dataset
df = pd.read_csv("../content/drive/MyDrive/student_performance.csv")

# 3. Preview data
df.head()
```

```python
# 5. Check missing values
df.isnull().sum()
```

```python
# 6. Handle missing values
df.fillna({
    'study_time': df['study_time'].median(),
    'attendance': df['attendance'].median()
}, inplace=True)
```

```python
# 7. Encode categorical variables
df['gender'] = df['gender'].map({'M': 0, 'F': 1})
df['parental_education'] = df['parental_education'].astype('category').cat.codes
```

```python
# 8. Save cleaned data
df.to_csv("../content/drive/MyDrive/cleaned_student_data.csv", index=False)

df.head()
```
